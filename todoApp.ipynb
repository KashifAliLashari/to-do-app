```python
# prompt: generate a simple todo app code in python

import json

def load_todos(filepath="todos.json"):
  """Loads todos from a JSON file."""
  try:
    with open(filepath, "r") as f:
      return json.load(f)
  except FileNotFoundError:
    return []

def save_todos(todos, filepath="todos.json"):
  """Saves todos to a JSON file."""
  with open(filepath, "w") as f:
    json.dump(todos, f, indent=4)

def add_todo(todos):
  """Adds a new todo to the list."""
  task = input("Enter new task: ")
  todos.append({"task": task, "completed": False})
  print("Task added successfully!")

def view_todos(todos):
  """Views all todos."""
  if not todos:
    print("No todos yet!")
    return

  for index, todo in enumerate(todos):
    status = "[x]" if todo["completed"] else "[ ]"
    print(f"{index + 1}. {status} {todo['task']}")

def complete_todo(todos):
    """Marks a todo as complete."""
    view_todos(todos)
    try:
        index = int(input("Enter the number of the task to mark as complete: ")) - 1
        if 0 <= index < len(todos):
            todos[index]["completed"] = True
            print("Task marked as complete!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a number.")


def remove_todo(todos):
    """Removes a todo from the list."""
    view_todos(todos)
    try:
        index = int(input("Enter the number of the task to remove: ")) - 1
        if 0 <= index < len(todos):
            del todos[index]
            print("Task removed successfully!")
        else:
            print("Invalid task number.")
    except ValueError:
        print("Invalid input. Please enter a number.")


def main():
  """Main function to run the todo app."""
  todos = load_todos()

  while True:
    print("\nChoose an action:")
    print("1. Add task")
    print("2. View tasks")
    print("3. Mark task as complete")
    print("4. Remove task")
    print("5. Exit")

    choice = input("Enter your choice (1-5): ")

    if choice == "1":
      add_todo(todos)
    elif choice == "2":
      view_todos(todos)
    elif choice == "3":
      complete_todo(todos)
    elif choice == "4":
        remove_todo(todos)
    elif choice == "5":
      save_todos(todos)
      print("Exiting...")
      break
    else:
      print("Invalid choice. Please try again.")

if __name__ == "__main__":
  main()
```
